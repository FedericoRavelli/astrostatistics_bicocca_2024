Create a Poisson distribution: 
                    P(k,mu) = [(mu)^k * exp(-mu)]/[k!]

where:
 > k = number of deaths respect to number of groups 
 
 > mu = avarage of the deaths number

```python
#len(N_deaths)
def f_poisson(k):
    y=np.zeros(len(k))
    mu = np.average(N_deaths, weights=N_groups)
    for i in range(len(k)):
        y[i]=((mu)**k[i])*np.exp(-mu)/math.factorial(k[i]) 
    return y
```

```python
print(f_poisson(N_deaths))
print(sum(f_poisson(N_deaths)))
```

```python
plt.plot(N_deaths,f_poisson(N_deaths))
plt.xlabel("Number of deaths")
plt.ylabel("f_poisson")
plt.title("Poisson distribution")
plt.show()
```
