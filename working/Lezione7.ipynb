```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
N = 1000
mu = 1
sigma = 0.02
chi = np.random.uniform(0,1,N)
M = np.random.normal(mu, sigma, N)
f = np.sqrt((1+np.sqrt(1-chi**2))/2) #f = M_irr/M
h = plt.hist(f, bins= 10,density = True)
```

```python
from astroML import stats as astroMLstats
sigmaG2 = astroMLstats.sigmaG(f)
print(sigmaG2)
```

```python
delta_bins= (2.7*sigmaG2)/(N**(1/3))
print(delta_bins)
```
