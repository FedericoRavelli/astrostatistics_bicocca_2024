# Compute $\pi$

I have forgotten the value of $\pi$, but:
- I know the formula for the area of a square 
- I know how to draw a circle.
 

```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

### a) 
- Drawing a square and circumscribing a circle in it.
- We out down random points within the square and note which ones land in the circle.

```python
# Draw a square with side equal to 1
Ntot = 10000
x = stats.uniform(0,1).rvs(Ntot)
y = stats.uniform(0,1).rvs(Ntot)

# Equation of circunference
square_radius = x**2 + y**2

plt.scatter(x,y, color = 'royalblue')
plt.scatter(x[square_radius<1 ],y[square_radius<1 ], color= 'red')
```

### b)
The ratio of random points in the circle to the number of random points drawn is releated to the area of our circle, allowing us to calculate $\pi$,
$\frac{Ncircle}{Ntot}.$

```python
Ncircle = np.sum(square_radius<1)
pi = 4* Ncircle/Ntot
print("Estimate of pi is = ",pi)
```
